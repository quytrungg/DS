**Câu 2 :** Ca sĩ sẽ chọn hát dòng nhạc nào phổ biến nhất

```python
data_2 = pd.DataFrame(pd_artist['genres'].str.split(r'\]|,|\[').explode().value_counts()).drop(['']).nlargest(n=10, columns='genres')
```

```python
data = data_2.to_numpy().reshape(-1)
labels = [i for i, _ in data_2.iterrows()]

colors = sns.color_palette('pastel')[0:10]

plt.pie(data, labels = labels, colors = colors, autopct='%.1f%%')
plt.show()
```

- Chúng ta thấy phần lớn các ca sĩ chọn cho mình dòng nhạc rock, pop, rap và các biến thể của chúng.
- Hơn 50% số lượng bài hát thuộc thể loại pop, rock và các biến thể của nó.
- Rap và pop rap chiếm lần lượt 9.4% và 9%, điều đó cho thấy thể loại nhạc rap hiện nay đang khá thịnh hành.
- Ngoài ra có loại nhạc 'electro house' cũng lọt vào top với 8.9%

## Cách tổ chức 

+ Dataframe `pd_track` mô tả thông tin về một playlist sẽ chứa các track nào, một `playlist_id` sẽ có thể có nhiều `track_id`. Ngoài ra còn một số thông tin khác ở đây chẳng hạn như độ phổ biến, thời lượng track,...
+ Sử dụng các `track_id` để truy xuất các thông tin đặc trưng về âm thanh tại dataframe `pd_audio`. Tại Dataframe này, các đặc trưng âm học sẽ được biểu diễn.

`pd_full` sẽ là một Dataframe tổng hợp tất cả thông tin bao gồm cả các đặc trưng về audio

```python
pd_track = pd.merge(pd_track, pd_audio, left_on = 'track_id', right_on = 'id')
pd_track

pd_full = pd.merge(pd_track, pd_playlist, left_on = 'playlist_id', right_on = 'playlist_id')
pd.set_option('display.max_columns', None)
pd_full
```

```python
pd_full = pd_full.drop('playlist_name', axis = 1)
pd_full = pd_full.drop('playlist_description', axis = 1)
pd_full = pd_full.drop('id', axis = 1)
```

```python
pd_full
```

```python
pd_full = pd_full.drop_duplicates()
pd_full = pd_full.fillna('None')
pd_full
```
