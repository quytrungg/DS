### Crawl a page with Selenium and parse it with Beautiful Soup

```python
# Include necesary packages
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from datetime import date, timedelta
```

```python
# Open Chrome browser
browser = webdriver.Chrome()
```

```python
# Open the homepage of Vietjet Air
browser.get('https://www.vietjetair.com/vi')
browser.maximize_window()
browser.implicitly_wait(20)
```

```python
# Dismiss the ads
browser.find_element(By.XPATH, "//*[text()='Để sau nhé']").click()
```

```python
# Choose the option of one-way ticket
browser.find_element(By.XPATH, "//*[text()='Một chiều']").click()
```

```python
# Click on the departure place box  
browser.find_element(By.XPATH, "//*[*/text()='Điểm khởi hành']").click()
```

```python
# Set the departure place to Sài Gòn
browser.find_element(By.XPATH, "//*[text()='SGN']").click()
```

```python
# Set the destination to Chu Lai
browser.find_element(By.XPATH, "//*[text()='VCL']").click()
```

```python
# Set the departure date to tomorrow
today = date.today()
tomorrow = today + timedelta(1)
print(tomorrow)
print(tomorrow.day, tomorrow.month)

elems = browser.find_elements(By.XPATH, f'//span[text()={tomorrow.day}]')

if today.month == tomorrow.month:
    elems[0].click()
else:
    elems[1].click()
```

```python
# Click on the button "Tìm chuyến bay"
browser.find_element(By.XPATH, "//button[*/*/text()='Tìm chuyến bay']").click()
```

```python
# Pass the page content to Beautiful Soup and end the mission of Selenium
html_text = browser.page_source
browser.quit()
tree = BeautifulSoup(html_text)
print(tree.prettify())
```
